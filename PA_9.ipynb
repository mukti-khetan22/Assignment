```python
# Function to check if a number is a Disarium Number
def is_disarium_number(number):
    str_number = str(number)
    sum = 0
    for i in range(len(str_number)):
        sum += int(str_number[i]) ** (i+1)
    return sum == number

# Function to print all Disarium Numbers between 1 and 100
def print_disarium_numbers():
    print("Disarium Numbers between 1 and 100:")
    for number in range(1, 101):
        if is_disarium_number(number):
            print(number)

# Function to check if a number is a Happy Number
def is_happy_number(number):
    def get_next_number(n):
        next_num = 0
        while n > 0:
            next_num += (n % 10) ** 2
            n //= 10
        return next_num

    seen = set()
    while number != 1 and number not in seen:
        seen.add(number)
        number = get_next_number(number)
    return number == 1

# Function to print all Happy Numbers between 1 and 100
def print_happy_numbers():
    print("Happy Numbers between 1 and 100:")
    for number in range(1, 101):
        if is_happy_number(number):
            print(number)

# Function to check if a number is a Harshad Number
def is_harshad_number(number):
    digits_sum = sum(int(digit) for digit in str(number))
    return number % digits_sum == 0

# Function to print all Harshad Numbers between 1 and 100
def print_harshad_numbers():
    print("Harshad Numbers between 1 and 100:")
    for number in range(1, 101):
        if is_harshad_number(number):
            print(number)

# Function to print all Pronic Numbers between 1 and 100
def print_pronic_numbers():
    print("Pronic Numbers between 1 and 100:")
    for number in range(1, 101):
        if number * (number + 1) <= 100:
            print(number * (number + 1))

# Main program
print_disarium_numbers()
print()
print_happy_numbers()
print()
print_harshad_numbers()
print()
print_pronic_numbers()
```
