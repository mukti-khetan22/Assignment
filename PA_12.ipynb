```python
from collections import OrderedDict

# 1. Function to extract unique values from dictionary values
def extract_unique_values(dictionary):
    unique_values = set()
    for values in dictionary.values():
        if isinstance(values, list):
            unique_values.update(values)
        else:
            unique_values.add(values)
    return list(unique_values)

# 2. Function to find the sum of all items in a dictionary
def find_sum_of_items(dictionary):
    return sum(dictionary.values())

# 3. Function to merge two dictionaries
def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()
    merged_dict.update(dict2)
    return merged_dict

# 4. Function to convert key-values list to a flat dictionary
def convert_to_flat_dictionary(key_value_list):
    flat_dict = {}
    for key, value in key_value_list:
        flat_dict[key] = value
    return flat_dict

# 5. Function to insert at the beginning in OrderedDict
def insert_at_beginning_ordered_dict(ordered_dict, key, value):
    ordered_dict[key] = value
    ordered_dict.move_to_end(key, last=False)

# 6. Function to check the order of characters in a string using OrderedDict
def check_order_of_characters(string, pattern):
    ordered_dict = OrderedDict.fromkeys(string)
    j = 0
    for char in ordered_dict:
        if j >= len(pattern):
            break
        if char == pattern[j]:
            j += 1
    return j == len(pattern)

# 7. Function to sort Python dictionaries by key or value
def sort_dictionary(dictionary, by_key=True):
    if by_key:
        return OrderedDict(sorted(dictionary.items()))
    else:
        return OrderedDict(sorted(dictionary.items(), key=lambda x: x[1]))

# Main program
dictionary = {'a': [1, 2, 3], 'b': [2, 3, 4], 'c': [3, 4, 5]}
print("Unique values from dictionary values:", extract_unique_values(dictionary))

dictionary = {'a': 1, 'b': 2, 'c': 3}
print("Sum of all items in the dictionary:", find_sum_of_items(dictionary))

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
print("Merged dictionaries:", merge_dictionaries(dict1, dict2))

key_value_list = [('a', 1), ('b', 2), ('c', 3)]
print("Converted flat dictionary:", convert_to_flat_dictionary(key_value_list))

ordered_dict = OrderedDict([('a', 1), ('b', 2)])
insert_at_beginning_ordered_dict(ordered_dict, 'c', 3)
print("OrderedDict after insertion at the beginning:", ordered_dict)

string = "Hello World"
pattern = "loW"
print("Does the string follow the pattern?", check_order_of_characters(string, pattern))

dictionary = {'a': 2, 'c': 1, 'b': 3}
print("Sorted dictionary by key:", sort_dictionary(dictionary, by_key=True))
print("Sorted dictionary by value:", sort_dictionary(dictionary, by_key=False))
```
