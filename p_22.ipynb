```python
# 1. What is the result of the code, and explain?
X = 'iNeuron'
def func():
    print(X)

func()
# Output: 'iNeuron'
# Explanation: The function `func()` prints the value of the variable `X` defined outside the function.
# Since `X` is defined before the function and is accessible within the function's scope, it prints 'iNeuron'.

# 2. What is the result of the code, and explain?
X = 'iNeuron'
def func():
    X = 'NI!'

func()
print(X)
# Output: 'iNeuron'
# Explanation: The function `func()` assigns the value 'NI!' to a new local variable `X` within its scope.
# It does not affect the value of the global variable `X`. Therefore, when we print `X` after calling the function,
# it still holds the initial value 'iNeuron'.

# 3. What does this code print, and why?
X = 'iNeuron'
def func():
    X = 'NI'
    print(X)

func()
print(X)
# Output:
# NI
# iNeuron
# Explanation: Inside the function `func()`, the local variable `X` is assigned the value 'NI' and is printed.
# The global variable `X` remains unaffected. Therefore, when we print `X` after calling the function,
# it still holds the initial value 'iNeuron'.

# 4. What output does this code produce? Why?
X = 'iNeuron'
def func():
    global X
    X = 'NI'

func()
print(X)
# Output: 'NI'
# Explanation: The function `func()` uses the `global` keyword to indicate that the variable `X` is a global variable,
# not a local variable. It assigns the value 'NI' to the global variable `X`. Therefore, when we print `X` after
# calling the function, it reflects the updated value 'NI'.

# 5. What about this code—what’s the output, and why?
X = 'iNeuron'
def func():
    X = 'NI'
    def nested():
        print(X)
    nested()

func()
print(X)
# Output:
# NI
# iNeuron
# Explanation: Inside the nested function `nested()`, the variable `X` is not defined, so it looks for the variable
# in the nearest enclosing scope, which is the `func()` function. Therefore, it prints 'NI'. The global variable `X`
# remains unaffected, and when we print `X` after calling the function, it holds the initial value 'iNeuron'.

# 6. How about this code: what is its output in Python 3, and explain?
def func():
    X = 'NI'
    def nested():
        nonlocal X
        X = 'Spam'
    nested()
    print(X)

func()
# Output: 'Spam'
# Explanation: In Python 3, the `nonlocal` keyword is used to indicate that a variable is a nonlocal variable,
# which means it is not local to the current function but is defined in an enclosing function. Inside the `nested()`
# function, `nonlocal X` allows accessing and modifying the variable `X` defined in the `func()` function.
# Therefore, when we print `X` after calling the function, it holds the value 'Spam' assigned within the nested function.
```
