# 1. Create a list called years_list, starting with the year of your birth, and each year thereafter until the year of your fifth birthday.
birth_year = 1980
years_list = list(range(birth_year, birth_year + 6))
print("years_list:", years_list, "\n")

# 2. In which year in years_list was your third birthday? Remember, you were 0 years of age for your first year.
third_birthday_year = years_list[3]
print("Year of the third birthday:", third_birthday_year, "\n")

# 3. In the years list, which year were you the oldest?
oldest_year = max(years_list)
print("The year when you were the oldest:", oldest_year, "\n")

# 4. Make a list called things with these three strings as elements: "mozzarella", "cinderella", "salmonella".
things = ["mozzarella", "cinderella", "salmonella"]
print("things:", things, "\n")

# 5. Capitalize the element in things that refers to a person and then print the list. Did it change the element in the list?
things[1] = things[1].capitalize()
print("Updated things list:", things)
print("Yes, the element 'cinderella' changed to 'Cinderella'\n")

# 6. Make a surprise list with the elements "Groucho", "Chico", and "Harpo".
surprise = ["Groucho", "Chico", "Harpo"]
print("surprise:", surprise, "\n")

# 7. Lowercase the last element of the surprise list, reverse it, and then capitalize it.
last_element = surprise[-1]
last_element = last_element.lower()[::-1].capitalize()
surprise[-1] = last_element
print("Updated surprise list:", surprise, "\n")

# 8. Make an English-to-French dictionary called e2f and print it.
e2f = {"dog": "chien", "cat": "chat", "walrus": "morse"}
print("e2f:", e2f, "\n")

# 9. Write the French word for walrus in your three-word dictionary e2f.
french_word_for_walrus = e2f["walrus"]
print("French word for walrus:", french_word_for_walrus, "\n")

# 10. Make a French-to-English dictionary called f2e from e2f. Use the items method.
f2e = {value: key for key, value in e2f.items()}
print("f2e:", f2e, "\n")

# 11. Print the English version of the French word "chien" using f2e.

french_word = "chien"
english_translation = f2e[french_word]
print(f"The English version of '{french_word}' is '{english_translation}'\n")

# 12. Make and print a set of English words from the keys in e2f.

english_words_set = set(e2f.keys())
print("English words:", english_words_set, "\n")

# 13. Make a multilevel dictionary called life.

    'animals': {

        'octopi': {},

    },
    'plants': {},

top_level_keys = life.keys()

animals_keys = life['animals'].keys()
