```python
# 1. Function to find the sum of elements in a list
def sum_of_elements(lst):
    return sum(lst)

# 2. Function to multiply all numbers in a list
def multiply_numbers(lst):
    result = 1
    for num in lst:
        result *= num
    return result

# 3. Function to find the smallest number in a list
def find_smallest_number(lst):
    return min(lst)

# 4. Function to find the largest number in a list
def find_largest_number(lst):
    return max(lst)

# 5. Function to find the second largest number in a list
def find_second_largest_number(lst):
    sorted_lst = sorted(lst)
    return sorted_lst[-2]

# 6. Function to find the N largest elements from a list
def find_n_largest_elements(lst, n):
    sorted_lst = sorted(lst, reverse=True)
    return sorted_lst[:n]

# 7. Function to print even numbers in a list
def print_even_numbers(lst):
    even_numbers = [num for num in lst if num % 2 == 0]
    print("Even numbers:", even_numbers)

# 8. Function to print odd numbers in a list
def print_odd_numbers(lst):
    odd_numbers = [num for num in lst if num % 2 != 0]
    print("Odd numbers:", odd_numbers)

# 9. Function to remove empty lists from a list
def remove_empty_lists(lst):
    return [sublist for sublist in lst if sublist]

# 10. Function to clone or copy a list
def clone_list(lst):
    return lst.copy()

# 11. Function to count occurrences of an element in a list
def count_occurrences(lst, element):
    return lst.count(element)

# Main program
my_list = [2, 4, 6, 8, 10]
print("Sum of elements:", sum_of_elements(my_list))
print("Product of numbers:", multiply_numbers(my_list))
print("Smallest number:", find_smallest_number(my_list))
print("Largest number:", find_largest_number(my_list))
print("Second largest number:", find_second_largest_number(my_list))
print("N largest elements:", find_n_largest_elements(my_list, 3))
print_even_numbers(my_list)
print_odd_numbers(my_list)

list_with_empty = [1, [], 3, [], 5]
print("List with empty lists:", list_with_empty)
print("List without empty lists:", remove_empty_lists(list_with_empty))

original_list = [1, 2, 3, 4, 5]
cloned_list = clone_list(original_list)
print("Original list:", original_list)
print("Cloned list:", cloned_list)

element_list = [1, 2, 3, 4, 2, 5, 2]
element = 2
print("Occurrences of", element, "in the list:", count_occurrences(element_list, element))
```
